```python
import urllib.request

file_url = "https://drive.google.com/uc?export=download&id=1nOCPWWYPJkCHQ9rUaGpSs-e26q5Otdtf"
output_file = "downloaded_file.wav"

# Create a request to the URL
request = urllib.request.Request(file_url, headers={'User-Agent': 'Mozilla/5.0'})

# Open the URL and download the file
with urllib.request.urlopen(request) as response, open(output_file, 'wb') as output:
    output.write(response.read())

print(f"File downloaded as {output_file}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display

# Parameters
sampling_rate = 8000
duration = 1.0
frequency = 440
dynamic_range = 0.5
amplitude_zero_duration = 0.25  # 250ms
amplitude_zero_samples = int(amplitude_zero_duration * sampling_rate)

# Time vector
t = np.arange(0, duration, 1/sampling_rate)

# Generate cosine signal
signal = dynamic_range * np.cos(2 * np.pi * frequency * t)

# Set amplitude to zero for the specified duration
signal[:amplitude_zero_samples] = 0

# Plot the signal
plt.plot(t, signal)
plt.title('Periodic Cosine Signal with Notch')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Display playback widget
display(Audio(signal, rate=sampling_rate))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters
duration = 1  # in seconds
sampling_rate = 8000
frequency = 440  # in Hz
amplitude_range = [-0.5, 0.5]

# Generate time array
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sine wave signal with dynamic range and fix amplitude to zero for the first 200 samples
sig = amplitude_range[1] * np.sin(2 * np.pi * frequency * t)
sig[:200] = 0  # Fix amplitude to zero for the first 200 samples

# Plot the signal
plt.plot(t, sig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave Signal - 440 Hz with Dynamic Range and Fixed Amplitude')
plt.grid(True)
plt.show()
```
